# Convert TFLite to TensorFlow.js (Colab Version)

Since the local environment (Windows + Python 3.12) is incompatible with `tensorflowjs` converter requirements, use this notebook in **Google Colab** to perform the conversion.

**Steps:**
1. Open [Google Colab](https://colab.research.google.com/).
2. File > Upload Notebook > Select this file.
3. Run the cells below.

```python
# 1. Install Converter
!pip install tensorflowjs==4.17.0
```

```python
# 2. Upload your 'wakeword_sequence.h5' file
# NOTE: You cannot convert .tflite to tfjs directly. You must use the Keras (.h5) model.
# If you don't have it, re-run 'python training/train_sequence.py' (ensure it saves .h5).

from google.colab import files
import os

print("Please upload 'wakeword_sequence.h5' when prompted.")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
print(f"Uploaded: {filename}")
```

```python
# 3. Run Conversion (Keras -> TFJS)
!mkdir -p model_tfjs
!tensorflowjs_converter --input_format=keras --output_format=tfjs_graph_model "{filename}" model_tfjs

print("Conversion complete.")
```

```python
# 4. Download Results
!zip -r model_tfjs.zip model_tfjs
files.download('model_tfjs.zip')
```
